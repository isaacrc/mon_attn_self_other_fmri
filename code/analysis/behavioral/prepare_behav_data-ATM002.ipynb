```python
def clean_file(data):
    df = pd.DataFrame(columns = ['Index', 'blk_name']) 
    df2 = data.iloc[:,5:]
    df1 = data.iloc[:,:5]

    for col in range(len(df1.columns)):
        for inx, a in enumerate(df1.iloc[0:,col]):
            if a != -1:
                df = df.append({'Index' : inx, 'blk_name' : df1.columns[col]},  
                ignore_index = True) 

    df2['Index'] = range(len(df2))
    new = pd.merge(df, df2, on=["Index"])
    new.sort_values(by=['Index'])
    return(new)
                
```

```python
## Expand / Label TRs
"""
0 = SM
1 = SC
2 = OM
3 = OC
4 = Re
requires list of labels ouputed by psychopy (column 1 - MM_self_title.started, etc.)
returns label list (order is preserved) and TR labels
"""

def label_lists(label):
    b = [[]]
    a = []
    for i in label:
        # substring label in psychopy output
        if i[1:4] == "M_s":
            a.append("SM")
            b.append([0]*num_tr)
        elif i[1:4] == "C_s":
            a.append("SC")
            b.append([1]*num_tr)        
        elif i[1:4] == "M_o":
            a.append("OM")
            b.append([2]*num_tr)
        elif i[1:4] == "C_o":
            a.append("OC")
            b.append([3]*num_tr)     
        else:
            a.append("Re")
            b.append([4]*num_tr)     
    return a, b[1:]
 

```

# Find button presses / Counts 

```python
## Reformats the button press array and returns the button press instances  ##
def org_bpress(arg):
    tem = len(arg.split())
    return [1] * tem
    
    
```

```python
## Reformats the time array and returns the rounded (two decimals) time stamp  ##
## Timestamp represents the time the button was pressed, start of the run is t-zero ## 
import re
def org_tpress(arg):
    tem = arg.split()
    a=[]
    for i in tem:
        S = i
        pattern = re.compile(r'\-?\d+\.\d+')
        weights = list(map(float, re.findall(pattern, S)))
        a.append(round(float(weights[0]),2))
    return(a)
    
    
```

```python
## Returns the TR prior to the button press ## 
def tr_stam(arg):
    a = []
    for i in arg:
        a.append(int(round(i / 1.5,0)))
    return(a)
```

```python
## Returns a list of 209 TRs, 1 is a button press, 0 is not ## 
def all_TR_pres(arg):
    a = []
    for i in range(1,210):
        if i in arg:
            a.append(1)
        else:
            a.append(0)
    return a
            
```

```python
# Take in raw column data and return vector 0s and 1s, button press or no button press
def run_press(pressN, timeN):
    n_press = org_bpress(pressN)
    ts_stamp = org_tpress(timeN)
    tr_stamp = tr_stam(ts_stamp)
    return all_TR_pres(tr_stamp)
```
